```python
import pandas as pd

x_df = pd.DataFrame(variables['LRtrain'])
x_df
```

```python
y_df = pd.DataFrame(variables['LRtraintgt'])
y_df
```

```python
w0 = np.full(x_df.shape[0],0.01)
b0 = 0

w0
```

```python
x_npy = x_df.to_numpy()
x_npy.shape
```

```python
y_npy = y_df[0].to_numpy()
y_npy
```

```python
def scale(array):
    minimum = min(array)
    array_range = max(array)-min(array)
    for i in range(len(array)):
        array[i] = (array[i] - minimum)/array_range

    return array
```

```python
x_df.shape[1]
```

```python
scale(y_npy)
y_npy
```

```python
print(x_npy)
np.apply_along_axis(scale, 1, x_npy)
```

```python
def y_pred(X,W,b):
    y = b
    for i in range(len(X)):
        y += X[i] * W[i]
    return y
```

```python
def dw(w,b,w_index,alpha):
    s = 0
    for i in range(x_df.shape[1]):
        X = x_npy[:, i]
        y = y_npy[i]
        y_predicted = y_pred(X,w,b)
        s += (y-y_predicted)*X[w_index]

    return alpha*s/x_df.shape[1]

def db(w,b,alpha):
    s = 0
    for i in range(x_df.shape[1]):
        X = x_npy[:, i]
        y = y_npy[i]
        y_predicted = y_pred(X,w,b)
        s += y - y_predicted

    return alpha*s/x_df.shape[1]
```

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(2)
plt.ion()
```

```python
errors = np.zeros(x_df.shape[1])

def cost(w,b):
    s = 0
    for i in range(x_df.shape[1]):
        X = x_npy[:, i]
        y = y_npy[i]
        y_predicted = y_pred(X,w,b)
        errors[i] = y - y_predicted
        s += (y - y_predicted)**2

    counts, bins = np.histogram(errors)
    axs[0].stairs(counts, bins)
    axs[0].set_xlabel("Error")
    axs[0].set_ylabel("Frequency")

    return s/2/x_df.shape[1]
```

```python

costs = []
steps = 30
lr = 0.01

for i in range(steps):
    costs.append(cost(w0,b0))

    axs[1].plot(range(i+1),costs)
    axs[1].set_xlabel("Step")
    axs[1].set_ylabel("Cost")

    print(f'step: {i}\tcost: {cost(w0,b0)}\tb: {b0}\tw: {w0[:5]}...')
    
    b_descent = db(w0,b0,lr)
    w_descents = np.zeros(x_df.shape[0])
    
    for w_index in range(x_df.shape[0]):
        w_descents[w_index] = dw(w0,b0,w_index,lr)
    
    w0 = np.add(w_descents,w0)
    b0 += b_descent
```

```python
storage = open("values.txt", "a")
storage.write(str(b0) + '\n')
for w in w0:
    storage.write(str(w) + '\n')
storage.close()
```
